```python
# Assignment
# -------------
# caesar cipher
# ------------------  + 3
# A B C D E F G H I J     Y Z
# 0 1 2 3 4 5 6 7
# D E F


# Ex: egg => hjj
#    bindu => elqg
#    Yash  => bd..

# ASSUMPTION: Ignore case -sensitivity
# HINTS: ord(), chr(), %


text = input("Enter a string: ").lower()

result = ""

for ch in text:
    if ch.isalpha():
        new_char = chr((ord(ch) - ord('a') + 3) % 26 + ord('a'))
        result += new_char
    else:
        result += ch

print("Encrypted String:", result)
```

### byte strings

###### bytes objects actually behave like
###### immutable sequences of integers, with each
###### value in the sequence restricted such that
###### 0 <= x < 256

```python
my_string = "Python is interesting."
print(my_string, type(my_string))
```

```python
# string with encoding 'utf-8'
b_string = bytes(my_string, "utf-8")
print(b_string, type(b_string))
```
