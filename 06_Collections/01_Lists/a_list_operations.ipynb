###
List Attributes

```python

numbers = [88, 99, 666]

print("type(numbers)    =", type(numbers))
print("dir(numbers)     =", dir(numbers))
```

```python

print("len(numbers)            =", len(numbers))
print("numbers.__len__()       =", numbers.__len__())
assert len(numbers) == numbers.__len__()
```

```python
print("str(numbers)           =", str(numbers))
print("type(str(numbers))     =", type(str(numbers)))
print("numbers.__str__()      =", numbers.__str__())
print("type(numbers.__str__())=", type(numbers.__str__()))
assert str(numbers) == numbers.__str__()
```

```python

# List Repetion operation
print("numbers * 3         =", numbers * 3)  # original object not modified
print("numbers             =", numbers)
# original object not modified
print("numbers.__mul__(3)  =", numbers.__mul__(3))
print("numbers             =", numbers)
assert numbers * 3 == numbers.__mul__(3)
```

```python

# original object IS modified
print("numbers.__imul__(3) =", numbers.__imul__(3))  # numbers *= 3
print("numbers             =", numbers)
```

```python

print("id(numbers)          =", id(numbers))
# object overwriting
numbers = [88, 99, 666]
print("id(numbers)          =", id(numbers))
numbers = [88, 99, 666]
print("id(numbers)          =", id(numbers))
```

```python
# NOTE: overwriting will everytime change the address of the object, in datastcurure

# Assignment: try checking the overwriting address change in all data types


# Integer
num = 10
print(f"\nInteger")
print(f"Before: {num = }, id = {id(num)}")
num = 20
print(f"After : {num = }, id = {id(num)}")

# Float
flt = 10.5
print(f"\nFloat")
print(f"Before: {flt = }, id = {id(flt)}")
flt = 20.5
print(f"After : {flt = }, id = {id(flt)}")

# String
text = "Hello"
print(f"\nString")
print(f"Before: {text = }, id = {id(text)}")
text = "Python"
print(f"After : {text = }, id = {id(text)}")

# Boolean
flag = True
print(f"\nBoolean")
print(f"Before: {flag = }, id = {id(flag)}")
flag = False
print(f"After : {flag = }, id = {id(flag)}")

# List
ml = [1, 2, 3]
print(f"\nList")
print(f"Before: {ml = }, id = {id(ml)}")
ml = [4, 5, 6]
print(f"After : {ml = }, id = {id(ml)}")

# Tuple
tp = (1, 2, 3)
print(f"\nTuple")
print(f"Before: {tp = }, id = {id(tp)}")
tp = (4, 5, 6)
print(f"After : {tp = }, id = {id(tp)}")

# Set
st = {1, 2, 3}
print(f"\nSet")
print(f"Before: {st = }, id = {id(st)}")
st = {4, 5, 6}
print(f"After : {st = }, id = {id(st)}")

# Dictionary
d = {"a": 1}
print(f"\nDictionary")
print(f"Before: {d = }, id = {id(d)}")
d = {"b": 2}
print(f"After : {d = }, id = {id(d)}")
```

```python
# Assignment - how to remove the last occurence of a element in the list
ml = [10, 20, 30, 20, 40, 20, 50]

element = 20

for i in range(len(ml) - 1, -1, -1):
    if ml[i] == element:
        ml.pop(i)
        break

print(ml)
```
