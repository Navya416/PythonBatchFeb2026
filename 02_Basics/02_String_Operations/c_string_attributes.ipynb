```python
# Assignment : identify few examples in diffenrent languages for casefold attribute usecase
"""
Assignment: Examples of casefold() in different languages
"""

# English
print("HELLO".casefold() == "hello".casefold())      # True

# German (ß -> ss)
print("Straße".casefold())
print("STRASSE".casefold())
print("Straße".casefold() == "STRASSE".casefold())   # True

# Turkish
print("İSTANBUL".casefold())
print("istanbul".casefold())

# Greek
print("ΟΣ".casefold())
print("ος".casefold())
print("ΟΣ".casefold() == "ος".casefold())            # True

# French
print("ÉCOLE".casefold())
print("école".casefold())
print("ÉCOLE".casefold() == "école".casefold())      # True

# Spanish
print("NIÑO".casefold())
print("niño".casefold())
print("NIÑO".casefold() == "niño".casefold())        # True
```

```python
# partition - will result in 3 values in tuple
print("Python Production".split("r"))       # ['Python P', 'oduction']
print("Python Production".partition("r"))   # ('Python P', 'r', 'oduction')
```
