```python
"""
Assignment:
Difference between dict['key'], dict.get(), and dict.setdefault()
"""

student = {
    "name": "Navya",
    "age": 24
}

print(f"Original Dictionary: {student}")

print("\n1. Using dict['key']")
print(f"student['name'] = {student['name']}")

try:
    print(student["city"])
except KeyError as err:
    print(f"KeyError: {err}")

print("\n2. Using dict.get()")
print(f"student.get('name') = {student.get('name')}")
print(f"student.get('city') = {student.get('city')}")
print(f"student.get('city', 'Hyderabad') = {student.get('city', 'Hyderabad')}")

print("\nDictionary after get():")
print(student)

print("\n3. Using dict.setdefault()")
print(f"student.setdefault('name') = {student.setdefault('name')}")
print(f"student.setdefault('city', 'Hyderabad') = {student.setdefault('city', 'Hyderabad')}")

print("\nDictionary after setdefault():")
print(student)
```

```python

# Assignment: del dict['key'] vs dict.pop('key')

student = {
    "name": "Navya",
    "age": 24,
    "city": "Hyderabad"
}

print(f"Original Dictionary: {student}")

# 1. Using del
print("\n1. Using del")

student1 = student.copy()

del student1["city"]      
print(f"After del: {student1}")

try:
    del student1["country"]   
except KeyError as err:
    print(f"KeyError: {err}")

# 2. Using pop()
print("\n2. Using pop()")

student2 = student.copy()

removed_value = student2.pop("city")   

print(f"Removed Value = {removed_value}")
print(f"After pop(): {student2}")

try:
    student2.pop("country")   
except KeyError as err:
    print(f"KeyError: {err}")

# pop() with default value
value = student2.pop("country", "Key Not Found")
print(f"pop('country', 'Key Not Found') = {value}")

print(f"Final Dictionary: {student2}")
```

```python
# dict replicas

animal = {"name": "zebra", "continent": ("Africa", "Asia")}
print(animal)
```

```python

# replicas - To get the keys of existing dictionary to new dict
animal_schema = {}
animal_schema = animal_schema.fromkeys(animal)
print(animal_schema)
# NOTE: Default value is None
```

```python
animal_schema = animal_schema.fromkeys(animal, None)
print(animal_schema)

animal_schema = animal_schema.fromkeys(animal, "None")
print(animal_schema)

animal_schema = animal_schema.fromkeys(animal, "")
print(animal_schema)
```

```python
# To flush the content in the dictionary
print()
print(f"{id(animal) = }")
animal.clear()  # will flush the key-value pairs in dict
print(f"After dict.clear(), {animal =}")
print(f"{id(animal) = }")
```

```python
# Zip() function

# ways of creating dictionary
a = {"one": 1, "two": 2, "three": 3}
b = dict(one=1, two=2, three=3)

c = dict([("two", 2), ("one", 1), ("three", 3)])
d = dict(zip(["one", "two", "three"], [1, 2, 3]))

e = dict({"three": 3, "one": 1, "two": 2})

assert a == b == c == d == e
```

```python
## Assignment: create a logic to interchange the keys with values, and vice versa, for dictionaries

student = {
    "name": "Navya",
    "age": 24,
    "city": "Hyderabad"
}

print(f"Original Dictionary : {student}")

# Interchange keys and values
swapped_dict = {}

for key, value in student.items():
    swapped_dict[value] = key

print(f"Swapped Dictionary  : {swapped_dict}")
```
